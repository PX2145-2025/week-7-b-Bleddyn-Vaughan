```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import random, seed
```

```python
#1
def walk(n_steps):
    steps = np.random.choice([-1, 1], size=n_steps) #random + or - 1
    position = np.cumsum(steps) #cumulative sum 
    return position 

n_steps=1000
random_walk=walk(n_steps) #puts the 1000 walks throught the function to get left or right positions 
plt.figure(figsize=(10, 5))
plt.plot(random_walk,color="black")
plt.title("1D Random Walk 1000 steps")
plt.xlabel("Step Number")
plt.ylabel("Position")
plt.grid(True)
plt.show()

n_walkers = 100

plt.figure(figsize=(10, 5))
for i in range(n_walkers):
    randomwalk_100 = walk(n_steps)
    plt.plot(randomwalk_100, alpha=0.5)

#plot 1000 steps
plt.title("1D Random Walk 1000 steps(10 walkers)")
plt.xlabel("Step Number")
plt.ylabel("Position")
plt.grid(True)
plt.show()

```

```python
#2
steps=1000
N=10

paticles=np.zeros((steps,N),int)
random_steps = np.random.choice([-1, 1], size=(steps, N))# random movement left or right 1
particles = np.cumsum(random_steps, axis=0)#cumalative sum
plt.figure(figsize=(10, 5))
for i in range(N):
    plt.plot(particles[:, i], alpha=0.7)
plt.title("1D Random Walks for 10 Walkers (1000 Steps)")
plt.xlabel("Step Number")
plt.ylabel("Position")
plt.grid(True)
plt.show()
```

```python
N = 500
particles = np.zeros((steps, N), int)
random_steps = np.random.choice([-1, 1], size=(steps, N))# random step left or right 1
particles = np.cumsum(random_steps, axis=0)#cumalative sum
mean_position = np.mean(particles, axis=1)#average of all walkers          
mean_squared_position = np.mean(particles**2, axis=1)#average of squares

plt.figure(figsize=(10, 5))
plt.plot(mean_position, color='red')
plt.title("Mean Position vs Step Number (500 Walkers)")
plt.xlabel("Step Number")
plt.ylabel("Mean Position")
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(mean_squared_position, color='green')
plt.title("Mean Squared Position vs Step Number (500 Walkers)")
plt.xlabel("Step Number")
plt.ylabel("Mean Squared Position")
plt.grid(True)
plt.show()
```

```python
#mean position is more random it movement is symetric equal chance of left and right where as the mean squared graph shows that the mean squared position increases linealry with the step number 
```

```python
#3
steps = 1000
N = 500
random_steps = np.random.choice([-1, 1], size=(steps, N))#random left or right movement 
positions = np.cumsum(random_steps, axis=0)#cumalative sum
#function to help plot the histograms
def plot_histogram(data, step_number, bins=30):
    plt.figure(figsize=(8, 5))
    plt.hist(data, bins=bins, color='skyblue', edgecolor='black')
    plt.title(f"Histogram of Walker Positions at diffrent step numbers ")
    plt.xlabel("Position")
    plt.ylabel("Number of Walkers")
    plt.grid(True, alpha=0.3)

plot_histogram(positions[-1, :], step_number=steps)#last step number 
plt.savefig("histogram at last stepnumber")

for step in [10, 100, 500]:
    plot_histogram(positions[step, :], step_number=step) #histogram of intermediate step numbers 
plt.savefig("histograms at step number 10,100,500")
```
